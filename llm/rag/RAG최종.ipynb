# 매물 (536개)

```python
from haversine import haversine
import pandas as pd
import json
```

```python
comb=pd.read_excel('data/송파구편의시설.xlsx',sheet_name='송파구편의시설')
```

```python
with open('data/maemul_clear.json') as maemul :
    maemul = json.load(maemul)
```

```python
combs = list(comb.itertuples(name=None, index=None))
```

```python
chunk = []
metadatas = []
for m in maemul :
    s = []
    vusdml = []
    tag = []
    for c in combs :
        if c[5] == '건강' :
            if haversine(c[2:4], (m['위도'], m['경도'])) <= 1.3 :
                vusdml.append(f"[{c[5]}] {c[0]}")
                tag.append(c[5])
        elif c[5] == '교육' :
            if haversine(c[2:4], (m['위도'], m['경도'])) <= 0.5 :
                vusdml.append(f"[{c[5]}] {c[0]}")
                tag.append(c[5])
        elif c[4] == '공원' :
            if haversine(c[2:4], (m['위도'], m['경도'])) <= 0.3 :
                vusdml.append(f"[공원] {c[0]}")
                tag.append("공원")
        elif c[4] in ('노인복지', '문화', '체육') :
            if haversine(c[2:4], (m['위도'], m['경도'])) <= 0.5 :
                vusdml.append(f"[{c[5]}] {c[0]}")
                tag.append(c[5])
        elif c[5] == '편의' :
            if haversine(c[2:4], (m['위도'], m['경도'])) <= 1 :
                vusdml.append(f"[쇼핑] {c[0]}")
                tag.append("쇼핑")
        elif c[4] == '교통' :
            if haversine(c[2:4], (m['위도'], m['경도'])) <= 0.25 :
                vusdml.append(f"[{c[5]}] {c[0]}")
                tag.append(c[5])
        
    if len(vusdml) > 0 :
        s.append(f"이름 : {m['이름']}")
        s.append(f"유형 : {m['유형']}")
        s.append(f"편의시설 : {', '.join(sorted(vusdml))}")
        
    
    if len(s) > 0 :
        chunk.append("\n".join(s))
        metadatas.append({"for" : "추천",
                          "이름" : m['이름'],
                          "편의시설" : ','.join((sorted(set(tag))))})
```

```python
len(chunk)
```

```python
vectorDB.add_texts(
    texts=chunk,
    metadatas=metadatas
)
```

### 메타데이터 테스트

```python
mmr_docs = vectorDB.max_marginal_relevance_search('난 8호선을 타고 출근하고 한 달에 한 번 병원에 가', fetch_k=50, k=3, lambda_mult=0.1, filter= {'for':'추천'})

print('객체 수 :', len(mmr_docs))
print(mmr_docs[0])
print(mmr_docs[1])
print(mmr_docs[2])
```

```python
chat_model = ChatOpenAI(model='gpt-3.5-turbo',
                        api_key=MY_API_KEY)
```

```python
my_retriever = vectorDB.as_retriever(search_type = 'mmr',
                                         search_kwargs = {'fetch_k':10, 'k':3, 'lambda_mult' :0.1})
```

```python
R_QA = RetrievalQA.from_chain_type(llm=chat_model,
                                    chain_type = 'stuff',
                                    retriever = my_retriever,
                                    return_source_documents = True)
```

```python
user_question = "전세 만기가 지났는데, 새로운 세입자를 구하지 못해서 못 나가고 있습니다. 그냥 집주인에게 오늘 시점으로 3개월 후 나가겠다고 하려고 하는데, 전화 대신 메시지로 의견 전달해도 될까요?3개월 후 시점으로 무조건 집 구해서 나가니까, 3개월 지난 날짜로 전세금 돌려달라고...만약 집주인이 아무 답변이 없으면 어떻게 해야 하나요"
```

```python
llm_response = R_QA.invoke(user_question)
print(llm_response['result'])
```
